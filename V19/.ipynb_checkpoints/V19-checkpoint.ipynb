```python
# theoretical functions
def gauss(x,m,s,a):
    return a * 1/np.sqrt(2*np.pi*s**2)*np.exp(-(x-m)**2/(2*s**2))

# gives index of last value higher than border
def last_higher(data,border):
    return np.where(data[data > border][-1] == data)[0][-1]

def peak_ev(data,maxPeakWidth,i,area=5):
    data = np.pad(data,(0,maxPeakWidth//2 + area),'constant')
    mean = (sum(data[i + maxPeakWidth//2 : i + maxPeakWidth//2 + area]) + sum(data[i - maxPeakWidth//2 - area : i - maxPeakWidth//2]))/(2*area)
    return (data[i] - mean)/np.sqrt(max(1,mean))

def gauss_fit(data,Δdata=0,pos=-1,n=1,Iwidth=400,maxPeakWidth=20,sigma_factor=3,fullOutput=False,lin=False):
    if (pos == -1):
        pos = len(data)//2
        Iwidth = len(data)+1
    interval = [max(0,pos - Iwidth//2),min(pos + Iwidth//2,len(data)-maxPeakWidth//2)]
    d = np.array([peak_ev(data,maxPeakWidth,i) for i in range(interval[0],interval[1])])
    p = np.zeros(n)
    if lin:
        fun = lambda x,m,s,a,b,c: gauss(x,m,s,a) * (b*x +c)
        opt = np.zeros([n,5])
        err = np.zeros([n,5])
        χ2_red = np.zeros([n])
        cov = np.zeros([n,5,5])
    
    else:
        fun = gauss
        opt = np.zeros([n,3])
        err = np.zeros([n,3])
        χ2_red = np.zeros([n])
        cov = np.zeros([n,3,3])
    
    for i in range(n):
        p[i] = np.argmax(d) + interval[0]
        if lin:
            p0 = (p[i]+1,1,data[p[i]],0,0)
        else:
            p0 = (p[i]+1,1,data[p[i]])
        d[max(0,np.argmax(d)-maxPeakWidth//2):min(np.argmax(d)+maxPeakWidth//2,len(d)-1)] = 1
        o,c = curve_fit(fun,ch,data,p0=p0)
        if (type(Δdata) == int and Δdata == 0):
            opt[i],cov[i] = curve_fit(fun,ch[max(0,np.floor(o[0] - max(5,sigma_factor * o[1]))):min(np.ceil(o[0] + max(5,sigma_factor * o[1])),len(data)-1)],data[max(0,np.floor(o[0] - max(5,sigma_factor * o[1]))):min(np.ceil(o[0] + max(5,sigma_factor * o[1])),len(data)-1)],p0=p0)
            err[i] = np.sqrt(np.diag(cov[i]))
        else:
            opt[i],cov[i] = curve_fit(fun,ch[max(0,np.floor(o[0] - max(5,sigma_factor * o[1]))):min(np.ceil(o[0] + max(5,sigma_factor * o[1])),len(data)-1)],data[max(0,np.floor(o[0] - max(5,sigma_factor * o[1]))):min(np.ceil(o[0] + max(5,sigma_factor * o[1])),len(data)-1)],p0=p0,sigma=Δdata[max(0,np.floor(o[0] - max(5,sigma_factor * o[1]))):min(np.ceil(o[0] + max(5,sigma_factor * o[1])),len(data)-1)],absolute_sigma=True)
            if not fullOutput:
                err[i] = np.sqrt(np.diag(cov[i]))
            else:
                print(o,p)
                χ2_red[i] = 1/ (-max(0.0,np.floor(o[0] - max(5.0,sigma_factor * o[1])))+min(np.ceil(o[0] + max(5.0,sigma_factor * o[1])),len(data)-1)-3) * sum([((data[j]-fun(ch[j],*opt[i]))/Δdata[j])**2 for j in range(max(0.0,np.floor(o[0] - max(5.0,sigma_factor * o[1]))).astype(np.int64),min(np.ceil(o[0] + max(5.0,sigma_factor * o[1])),len(data)-1).astype(np.int64))])
    if fullOutput:
        return [np.squeeze(opt),np.squeeze(χ2_red),np.squeeze(cov)]
    else:
        return [np.squeeze(opt),np.squeeze(err)]
```

### Efficiency of the detector

```python
days = 7891/365.25
Δdays = 1/365.25
t_m =np.array([300,300,300,400,400,350,350,200,300,300,300])
Δt_m =2
n_p =np.rollaxis(opt_f,1)[2]
Δn_p =np.rollaxis(err_f,1)[2]
a_old =np.array([411,407,407,429,429,400,400,394,477,477,477])
Δa_old =np.array([411*0.04,407*0.05,407*0.05,429*0.04,429*0.04,400*0.04,400*0.04,394*0.04,477*0.05,477*0.05,477*0.05])
t12 =np.array([30.17,10.5,10.5,5.272,5.272,2.602,2.602,432.6,13.33,13.33,13.33])
I = np.array([85,33,62,100,100,180,100,35,28,7,21])
a_n = a_old *np.exp(-np.log(2)*days/t12)
Δa_n =a_n*np.sqrt((Δa_old/a_old)**2 + (np.log(2)*Δdays/t12)**2)
eps = n_p /(a_n*I*t_m)
Δeps = eps* np.sqrt((Δa_n/a_n)**2 + (Δt_m/t_m)**2 + (Δn_p/n_p)**2)
```
